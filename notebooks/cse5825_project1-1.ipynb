***creating new features that I think make sense, including:***
- Likes / Likes + Dislikes Total (% likes) 
- Dislikes / Likes + Dislikes Total (% dislikes)
- Likes / Views (% of viewers who also like)
- Dislikes / Views (% of viewers who also dislike)
- Likes + Dislikes / Views (% of viewers who also interact)
- Comments / Views (% of viewers who also comment)

```python
USvideos["likes_to_reacts"] = USvideos["likes"]/(USvideos["likes"]+USvideos["dislikes"])
USvideos["dislikes_to_reacts"] = USvideos["dislikes"]/(USvideos["likes"]+USvideos["dislikes"])
USvideos["likes_to_views"] = USvideos["likes"]/USvideos["views"]
USvideos["dislikes_to_views"] = USvideos["dislikes"]/USvideos["views"]
USvideos["reacts_to_views"] = (USvideos["likes"] + USvideos["dislikes"])/USvideos["views"]
USvideos["comments_to_views"] = USvideos["comment_count"]/USvideos["views"]
```
